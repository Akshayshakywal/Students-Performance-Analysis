```python
gender_performance = data.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print("average score by gender:")
print(gender_performance)
```

```python
import matplotlib.pyplot as plt
```

```python
gender_performance.plot(kind='bar', figsize=(10, 6), color=['black', 'purple', 'green'])
plt.title('Average Scores by Gender in Math, Reading, and Writing')
plt.ylabel('Average Score')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.legend(title='Subjects')
plt.show()
```

```python
# Group data by lunch type and calculate the average scores
lunch_performance = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Group data by test preparation course and calculate the average scores
prep_course_performance = data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()

# Display the results
print("Average scores by lunch type:")
print(lunch_performance)
print("\nAverage scores by test preparation course completion:")
print(prep_course_performance)

# Visualize the results for lunch type
lunch_performance.plot(kind='bar', figsize=(10, 6), color=['blue', 'orange', 'green'])
plt.title('Average Scores by Lunch Type')
plt.ylabel('Average Score')
plt.xlabel('Lunch Type')
plt.xticks(rotation=0)
plt.legend(title='Subjects')
plt.show()

# Visualize the results for test preparation course
prep_course_performance.plot(kind='bar', figsize=(10, 6), color=['purple', 'pink', 'teal'])
plt.title('Average Scores by Test Preparation Course Completion')
plt.ylabel('Average Score')
plt.xlabel('Test Preparation Course')
plt.xticks(rotation=0)
plt.legend(title='Subjects')
plt.show()
```

```python
#task02
```

```python
# Grouping students by lunch type and converting each group into a list of dictionaries
students_by_lunch = data.groupby('lunch').apply(lambda x: x.to_dict('records')).to_dict()

# Grouping students by test preparation course and converting each group into a list of dictionaries
students_by_prep_course = data.groupby('test preparation course').apply(lambda x: x.to_dict('records')).to_dict()

# Calculating average scores by lunch type
avg_scores_by_lunch = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean().to_dict(orient='index')

# Calculating average scores by test preparation course
avg_scores_by_prep_course = data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().to_dict(orient='index')

# Displaying the results
print("Students grouped by lunch type:")
for lunch_type, students in students_by_lunch.items():
    print(f"\nLunch Type: {lunch_type} | Number of Students: {len(students)}")
    print(f"Average Scores: {avg_scores_by_lunch[lunch_type]}")

print("\nStudents grouped by test preparation course:")
for prep_course, students in students_by_prep_course.items():
    print(f"\nTest Preparation Course: {prep_course} | Number of Students: {len(students)}")
    print(f"Average Scores: {avg_scores_by_prep_course[prep_course]}")
```
